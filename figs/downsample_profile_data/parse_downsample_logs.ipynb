```python
import os
import numpy as np
import rpy2.robjects as robj

from rpy2.robjects import r
from rpy2.robjects.numpy2ri import numpy2ri
```

```python
fnames = [file for root, dirs, files in os.walk('./') for file in files if '.txt' in file]

dsf = np.array(())
roi = np.array(())
tat = np.array(())
tgp = np.array(())
trd =np.array(())
tat_e = np.array(())
tgp_e = np.array(())
trd_e =np.array(())

for fname in fnames:
    fl = open(fname)

    rois  = np.array(())
    t_atl = np.array(())
    t_grp = np.array(())
    t_rad = np.array(())

    for line in fl:
        if "Read took" in line:
            t_rad = np.append(t_rad, float(line.split('Read took ')[1].split(' sec')[0]))
        elif "Completed building graph" in line:
            t_grp = np.append(t_grp, float(line.split('graph in ')[1].split(' sec')[0]))
        elif "Your atlas has" in line:
            rois = np.append(rois, float(line.split('atlas has ')[1].split(' regions')[0]))
        elif "Building atlas took" in line:
            t_atl = np.append(t_atl, float(line.split('Building atlas took ')[1].split(' sec')[0]))
        elif "downsampling to" in line:
            continue
        else:
            print "What is this? ... ", line
            
    dsf = np.append(dsf, int(fname.split('.txt')[0].split('ds')[1]))

    assert (len(np.unique(rois))==1)
    roi = np.append(roi, np.unique(rois))

    tgp = np.append(tgp, np.mean(t_grp))
    tgp_e = np.append(tgp_e, np.std(t_grp)/np.sqrt(len(t_grp)))

    trd = np.append(trd, np.mean(t_rad))
    trd_e = np.append(trd_e, np.std(t_rad)/np.sqrt(len(t_rad)))

    tat = np.append(tat, np.mean(t_atl))
    tat_e = np.append(tat_e, np.std(t_atl)/np.sqrt(len(t_atl)))
    
    print "DS Factor:", dsf[-1]
    print "N:", np.unique((len(t_grp), len(t_atl), len(t_rad), len(rois))) 
    print "ROIs:", roi[-1]
    print "Time to make atlases:", tat[-1], "+/-", tat_e[-1]
    print "Time to read graph:", trd[-1], "+/-", trd_e[-1]
    print "Time to create graph:", tgp[-1], "+/-", tgp_e[-1]
    print "-------------"
    
dsfr = numpy2ri(dsf)
roir = numpy2ri(roi)
tgpr = numpy2ri(tgp)
trdr = numpy2ri(trd)
tatr = numpy2ri(tat)
tgp_er = numpy2ri(tgp_e)
trd_er = numpy2ri(trd_e)
tat_er = numpy2ri(tat_e)

r.assign('dsf', dsfr)
r.assign('roi', roir)
r.assign('tgp', tgpr)
r.assign('trd', trdr)
r.assign('tat', tatr)
r.assign('tgp_e', tgp_er)
r.assign('trd_e', trd_er)
r.assign('tat_e', tat_er)

r("save(dsf, roi, tgp, trd, tat, tgp_e, trd_e, tat_e, file='"+"ds_data.gzip"+"', compress=TRUE)")
```
